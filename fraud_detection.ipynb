```python
train_df_transformed = train_df.copy()
skewness_before = train_df[features].skew()

for col in features:
    if abs(train_df[col].skew()) > 0.75:
        train_df_transformed[col] = np.where(
            train_df[col] >= 0,
            np.log1p(train_df[col]),
            -np.log1p(-train_df[col])
        )

skewness_after = train_df_transformed[features].skew()
skewness_comparison = pd.DataFrame({
    'Skewness Before': skewness_before,
    'Skewness After': skewness_after
})
print(skewness_comparison)
```

```python
fig, axes = plt.subplots(10, 3, figsize=(15, 40))
axes = axes.flatten()
for i, feature in enumerate(features):
    sns.histplot(train_df_transformed[feature], ax=axes[i], kde=False, bins=30)
    axes[i].set_title(f'{feature} after Transformation')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')
for i in range(len(features), len(axes)):
    fig.delaxes(axes[i])
plt.tight_layout()
plt.show()
```

```python
X = train_df_transformed.drop("Class", axis=1)
y = train_df_transformed["Class"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
```

```python
rob_scaler = RobustScaler()
X_train[['Amount','Time']] = rob_scaler.fit_transform(X_train[['Amount','Time']])
X_test[['Amount','Time']] = rob_scaler.transform(X_test[['Amount','Time']])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=42, class_weight='balanced')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

print("Random Forest")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
cm = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Random Forest')
plt.show()
print(classification_report(y_test, y_pred_rf))
```
